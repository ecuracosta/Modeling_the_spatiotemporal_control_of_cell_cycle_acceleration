## Figure 2A

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('ggplot')
%matplotlib inline
```

```python
# Read simulations data and filters
heatmap = pd.read_csv("../Fitting/Switchpoint_fit.csv")
heatmap.index = heatmap['Unnamed: 0']
heatmap = heatmap.drop('Unnamed: 0',1)

fig = plt.figure()
ax = fig.add_subplot(111)

# Heatmap
ax = sns.heatmap(heatmap.astype(float), annot=False, fmt="g", cmap=sns.cubehelix_palette(90, start=2, rot=0.05, dark=0.2, light=1, reverse=True),xticklabels=40,yticklabels=50,cbar=True)
cbar = ax.collections[0].colorbar
cbar.set_label('Residual sum of squares', labelpad=32)
plt.ylabel(r'$N_0$')
plt.xlabel(r'$\lambda$' + ' (' + r'$\mu$'+'m)')
plt.suptitle('Parameters space', y=0.98, x=0.42, size=50)
fig.set_size_inches(24, 16)
plt.rcParams.update({'font.size': 34})
filename = "../../Plots/" + "Fig_2A.png"
fig.savefig(filename, dpi=300)
plt.show()
```
