# Model Shrink-Shrink

```python
import numpy as np
import os,sys
import pandas as pd
import time
import random
```

```python
def cellDivisions(tm,td,ts,stem,rl,positions,tv):
    global n
    for dividingCell in range(0,n):
        tc = ts - tm[dividingCell]
        tr = td[dividingCell] - tc
        if (dividingCell-n0)*13.2 > rl and ts == tv:    # Pulse
            M_G2 = np.random.normal(9,6)
            S_noSkip = np.random.normal(88,9)
            S_skip = np.random.normal(91,9)
            G1_noSkip = np.random.normal(22,19)
            G1_skip = np.random.normal(130,19)
            if tr > M_G2:    # si estoy en G2/M la dejo continuar
                if M_G2 <= tr < M_G2+S_noSkip+S_skip:     # en fase S
                    td[dividingCell] = M_G2+((tr-M_G2)*S_noSkip)/(S_noSkip+S_skip)
                    tm[dividingCell] = ts
                elif M_G2+S_noSkip+S_skip <= tr:          # en fase G1
                    td[dividingCell] = M_G2+S_noSkip+((tr-M_G2-S_noSkip-S_skip)*G1_noSkip)/(G1_noSkip+G1_skip)
                    tm[dividingCell] = ts
                else:
                    print("Not posible")
            stem[dividingCell] = True
        if tr < 1/3600:
            tm[dividingCell] = ts
            tm = np.insert(tm,dividingCell,ts)
            tm = tm[:-1]            
            n = n+1
            positions = np.insert(positions,dividingCell,-1)
            if stem[dividingCell]:
                td[dividingCell] = np.random.normal(119, 10)
                td = np.insert(td,dividingCell,np.random.normal(119, 10))
                td = td[:-1]
                stem = np.insert(stem,dividingCell,True)
                stem = stem[:-1]
            else:
                td[dividingCell] = np.random.normal(340, 32)
                td = np.insert(td,dividingCell,np.random.normal(340, 32))
                td = td[:-1]
                stem = np.insert(stem,dividingCell,False)
                stem = stem[:-1]                
    return tm,td,stem,positions
```

```python
def run(steps,tm,td,stem,rl,positions,tv):
    global n
    outgrowth = []
    tracking = []
    tracking.append(positions)
    for ts in range(0,steps+1,1):
        tm,td,stem,positions = cellDivisions(tm,td,ts,stem,rl,positions,tv)
        outgrowth.append(n-n0)    # Appends the outgrowth for each time
        tracking.append(positions)
    return outgrowth,tracking
```

```python
root = '/home/ecuracosta/Desktop/MS/Simulations/Models/Shrink-Shrink/'
os.chdir(root)

start_time = time.time()

random.seed(33)

emptySpace = 250
steps = 24*8  # Number of steaps (in hours)
n0 = 205  # Number of cells after amputation
rl = -1045  # Recruitment limit
tv_array = np.arange(0,8*24+1,1)

for tv in tv_array:
    lattice = n0+emptySpace
    folder = 'n0='+str(n0)+'_rl='+str(rl)+'_tv='+str(tv)
    if not os.path.isdir(root+folder):
        os.makedirs(folder)
        os.makedirs(folder+'/'+'Outgrowth')
        os.makedirs(folder+'/'+'Trajectories')
        print('n0:',n0)
        print('rl:',rl)
        print('tv:',tv)
        print('-------')
        for seed in range(0,200):
            n = n0
            positions = np.arange(0,n0) 
            tm = np.random.uniform(-(340-2*32),0,lattice)
            td = np.random.normal(340,32,lattice)
            stem = np.zeros(lattice)
            solution,track = run(steps,tm,td,stem,rl,positions,tv)

            d1 = {"time":np.arange(0,steps+1),"outgrowth":solution}
            df1 = pd.DataFrame(d1)
            outfile1 = open(str(root)+str(folder)+'/'+'Outgrowth'+'/'+'sim_seed='+str(seed)+'.csv', 'a')
            df1.to_csv(outfile1, sep=',')
            outfile1.close()
            
            d2 = {"positions":track}
            df2 = pd.DataFrame(d2)
            outfile2 = open(str(root)+str(folder)+'/'+'Trajectories'+'/'+'Trajectories_sim_seed='+str(seed)+'.csv', 'a')
            df2.to_csv(outfile2, sep=',')
            outfile2.close() 
            
print("Total time: %s seconds <--" % (time.time() - start_time))
```
