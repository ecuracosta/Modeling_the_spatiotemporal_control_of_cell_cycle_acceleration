```python
# required libraries
import os
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
%matplotlib inline
```

```python
switchpoint = np.asarray([-794, -641, -261])
switchpoint_time = np.asarray([4, 6, 8])
switchpoint_error = np.asarray([-130, -140, -187])

outgrowth = np.asarray([0, 0.0565, 0.16328571, 0.45175, 1.2785, 2.25725])*1000
outgrowth_time = np.asarray([0, 2, 3, 4, 6, 8])
```

```python
fig = plt.gcf()
ax = fig.add_subplot(111)

# Experimental Switchpoint
#ax.errorbar(switchpoint_time , switchpoint, yerr=switchpoint_error, color='black',ecolor='black', fmt='o',markersize=11,elinewidth=3)
ax.errorbar(switchpoint, switchpoint_time, xerr=switchpoint_error, color='black',ecolor='black', fmt='o',markersize=11,elinewidth=3)

# # Dark green region
# # ax.fill_between(outgrowth_time, -2000, outgrowth, facecolor='g', alpha=0.2)
ax.fill_between([-2000,0*1000, 0.0565*1000, 0.16328571*1000, 0.45175*1000, 1.2785*1000, 2.25725*1000],[0,0,2,3,4,6,8],8, facecolor='g', alpha=0.5)

# # Light green region
# # ax.fill_between([4,6,8], [-794, -641, -261], np.asarray([0.45175, 1.2785, 2.25725])*1000, facecolor='g', alpha=0.5)
ax.fill_between([-794, -641, -261, 0.45175*1000, 1.2785*1000, 2.25725*1000], [4,4,4,4,6,8], [4,6,8,8,8,8],facecolor='white', alpha=0.4)

# # Dashed region
# # ax.fill_between(outgrowth_time, outgrowth, 2500.0, facecolor='gray', alpha=0.2, hatch='/', edgecolor = 'black')
ax.fill_between(outgrowth, outgrowth_time, 0, facecolor='gray', alpha=0.2, hatch='/', edgecolor = 'black')

# # Dotted region
# # ax.fill_between([0, 8], 2 * [-2000], -3000, facecolor='gray', alpha=0.2, hatch='.', edgecolor = 'black')
ax.fill_between([-3000,0.45175*1000, 1.2785*1000, 2.25725*1000, -2000], 0, 8, facecolor='gray', alpha=0.2, hatch='.', edgecolor = 'black')

# # Text
# # ax.annotate('Long cell cycle zone', xy=(0.5,-1250),  xycoords='data', xytext=(1.2, -1250), size=24)
# # ax.annotate('Short cell cycle zone', xy=(4.5,100),  xycoords='data', xytext=(4.55,100), size=24)
ax.annotate('Long cell cycle zone', xy=(-1250,0.5),  xycoords='data', xytext=(-1500,2.5), size=24)
ax.annotate('Short cell cycle zone', xy=(100,4.5),  xycoords='data', xytext=(-200,6.5), size=24)

# # plt.xlabel('Time (days)')
# # plt.ylabel('AP position' + ' (' + r'$\mu$'+'m)')
# # plt.xlim(-0.1,8.1)
# # plt.ylim(-3000,2500)

plt.ylabel('Time (days)')
plt.xlabel('AP position' + ' (' + r'$\mu$'+'m)')
plt.ylim(-0.1,8.1)
plt.xlim(-3000,2500)
plt.gca().invert_yaxis()

#plt.title('Spacetime of cell proliferation \n during axolotl spinal cord regeneration', y=1.05, size=60)

filename = "./" + "Fig_S1A.png"

plt.rcParams.update({'font.size': 32})
fig.set_size_inches(18.5, 12.5)
fig.savefig(filename, dpi=300, bbox_inches='tight')
```
