```python
# required libraries
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
%matplotlib inline
```

```python
def plot_create():
    #plt.title('Prediction of spinal cord outgrowth', y=1.1, size=65)    
    exp_time = [0, 2, 3, 4, 6, 8]
    exp_outgrowth = [0, 60, 160, 450, 1280, 2260]
    exp_outgrowth_err = [0, 10, 20, 40, 60, 70]
    exp_time2 = [0, 1, 2, 3, 4, 5]
    
    # Experimental outgrowth
    ax.errorbar(exp_time , exp_outgrowth, yerr=exp_outgrowth_err, color='darkred',ecolor='darkred', fmt='s',markersize=20,elinewidth=3, label='Rodrigo-Albors $\it{et\ al}$., 2015')
    ax.errorbar(exp_time2 , outgrowth2_mean, yerr=outgrowth2_std, color='black',ecolor='black', fmt='o',markersize=20,elinewidth=3, label='This study')

    ax.legend(loc='upper left', facecolor='white')
    plt.xlabel('Time (days)')
    plt.ylabel('Outgrowth' + ' (' + r'$\mu$'+'m)')
    plt.xlim(-0.2,8.2)
    plt.ylim(-50,2500)
    fig.set_size_inches(18.5, 12.5)
    plt.rcParams.update({'font.size': 24})
    filename = './' + 'Fig_2B.png'    
    fig.savefig(filename, dpi=300, bbox_inches='tight')
```

```python
# Experiments 2
fig = plt.figure()
ax = fig.add_subplot(111)
outgrowth2_df = pd.read_csv('../data_analysis/G1S_switchpoint_fitting/'+'outgrowth.csv')
outgrowth2_df.set_index(['day', 'tail'], inplace=True)
outgrowth2_mean = outgrowth2_df.groupby('day').mean()['outgrowth']
outgrowth2_std = outgrowth2_df.groupby('day').std()['outgrowth']
plot_create()
```
