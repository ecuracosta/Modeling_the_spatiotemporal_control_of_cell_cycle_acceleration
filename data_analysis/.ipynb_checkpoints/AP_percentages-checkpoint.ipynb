```python
# Required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import scipy
import os
import seaborn as sns 
from pyabc import (Distribution, History)
```

```python
# Experiments
outgrowth_df = pd.read_csv('./'+'outgrowth.csv')
outgrowth_df.set_index(['day', 'tail'], inplace=True)
outgrowth_mean = outgrowth_df.groupby('day').mean()['outgrowth']

percentage_df = pd.read_csv('./'+'percentage_100um.csv')
df = percentage_df
for day in range(0,6):
    df.loc[df['day'] == day, 'position'] = (outgrowth_mean[day] - (df.loc[df['day'] == day, 'position']-100)).astype(int)
percentage_df = df
percentage_df.set_index(['day', 'tail', 'position'], inplace=True)
percentage_df = percentage_df.drop(['unlabelled'], axis=1)
experiments = percentage_df
```

```python
means,stds = {},{}
for day,df_day in percentage_df.groupby(level='day'):
        tails_mean, tails_std = {},{}
        for tail,df_animal in df_day.groupby(level='tail'):
            db_path = ("sqlite:///" + os.path.join("./fitting_results/",
                                               "sp_fitting-day="+str(day)+"-tail="+str(tail)+".db"))
            h = History(db_path)       
            df = h.get_distribution(m=0)
            tails_mean[tail] = df[0].mean()
            tails_std[tail] = df[0].std()         
        means[day] = pd.DataFrame.from_dict(tails_mean, orient='index')
        stds[day] = pd.DataFrame.from_dict(tails_std, orient='index')
means = pd.concat(means, names=['day','tail'])   
means['sp'] = outgrowth_mean-means['sp']
stds = pd.concat(stds, names=['day','tail'])   
```

```python
sp = {0:-863.0,1:-947.0,2:-754.0,3:-657.0,4:-718.0,5:-446.0}
```

```python
# Local fitting plots
dataGreen = means.drop('sp',axis=1).reset_index().drop(['c1m','c2m'],axis=1)
dataGreen.columns = ['day','tail','Posterior','Anterior']
dataGreen = dataGreen.melt(['day','tail'],var_name='cols', value_name='vals').sort_values('cols')
dataMagenta = means.drop('sp',axis=1).reset_index().drop(['c1g','c2g'],axis=1)
dataMagenta.columns = ['day','tail','Posterior','Anterior']
dataMagenta = dataMagenta.melt(['day','tail'],var_name='cols', value_name='vals').sort_values('cols')
for day,sp_mean in sp.items():
    dataG = dataGreen[dataGreen['day']==day]
    dataM = dataMagenta[dataMagenta['day']==day]
    ax = sns.scatterplot(x='cols', y='vals', data=dataG,color='green',edgecolor="white",linewidth=1,style='tail', s=100,legend=False)
    ax = sns.boxplot(x='cols', y='vals', data=dataG,color='green')
    ax = sns.scatterplot(x='cols', y='vals', data=dataM,color='magenta',edgecolor="white",linewidth=1,style='tail', s=100)
    ax2 = sns.boxplot(x='cols', y='vals', data=dataM,color='magenta')
    
    # Kolmogorov-Smirnov statistic
    significance = 0.05
    print("Day",day)
    green   = scipy.stats.ks_2samp(dataG[dataG['cols']=='Anterior']['vals'],dataG[dataG['cols']=='Posterior']['vals'])
    magenta = scipy.stats.ks_2samp(dataM[dataM['cols']=='Anterior']['vals'],dataM[dataM['cols']=='Posterior']['vals'])
    print("Green:",green)
    print("H0:",green[1]>significance)
    print("Magenta",magenta)
    print("H0:",magenta[1]>significance)    

    for i,artist in enumerate(ax2.artists):
        # Set the linecolor on the artist to the facecolor, and set the facecolor to None
        col = artist.get_facecolor()
        artist.set_edgecolor(col)
        artist.set_facecolor('None')

        # Each box has 6 associated Line2D objects (to make the whiskers, fliers, etc.)
        # Loop over them here, and use the same colour as above
        for j in range(i*6,i*6+6):
            line = ax2.lines[j]
            line.set_color(col)
            line.set_mfc(col)
            line.set_mec(col)
        
    title = 'Time = '+ str(day)+" dpa"
    plt.ylim(0,110)
    plt.xlabel('Relative position to the AP-border')
    plt.ylabel('G0/G1 and S/G2 AxFUCCI cells (%)')
    plt.suptitle(title,size='24')    
    plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
    plt.rcParams.update({'font.size': 14})
    plt.savefig('./boxes-day_'+str(day), dpi=300, bbox_inches='tight')
    plt.show()
```
