```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import scipy
import os
import seaborn as sns 
```

```python
# Experiments
outgrowth_df = pd.read_csv('./G1S_switchpoint_fitting/'+'outgrowth.csv')
outgrowth_df.set_index(['day', 'tail'], inplace=True)
outgrowth_mean = outgrowth_df.groupby('day').mean()['outgrowth']

percentage_df = pd.read_csv('./G1S_switchpoint_fitting/'+'percentage_100um.csv')
df = percentage_df
for day in range(0,6):
    df.loc[df['day'] == day, 'position'] = (outgrowth_mean[day] - (df.loc[df['day'] == day, 'position']-100)).astype(int)
percentage_df = df
percentage_df.set_index(['day', 'tail', 'position'], inplace=True)
percentage_df = percentage_df.drop(['unlabelled'], axis=1)
experiments = percentage_df
#experiments.drop(['double','mitosis'], axis=1)
```

```python
# Simulations

# LINEAR RECRUITMENT
n0_mean,n0_std = 196,2 # n0 mean and standar deviation
l_mean,l_std = 828,30      # lambda mean and standar deviation
tau_mean,tau_std = 85,12   # tau mean and standar deviation
root = "../main/simulations/"
model = 'outgrowth-TEST/'
parameters = 'n0='+str(n0_mean)+'\n'+'l='+str(l_mean)+'\n'+'tau='+str(tau_mean)+'/'

path = root+model+parameters
files = os.listdir(path)

all_seeds = []
intensity = {}
times = np.arange(0,6)
for seed in files:    
    all_times = []
    opened_file = open(path+seed)
    data = pd.read_csv(opened_file, delimiter=',')   
    data.set_index(['time', 'id'], inplace=True)
    intensity[seed] = data.xs(0,level='time').groupby('phase').count()['position']['S']   
    for day in times:
        data_time = data.xs(day*24,level='time')
        data_time = data_time.drop(['td','tc','recruited','dormant'], axis=1).dropna()
        
        exp_bins = percentage_df.xs(day,level='day').groupby('position').mean().index
        bins = (np.arange(exp_bins.min()-50,exp_bins.max()+150,100)).astype(int)
        data_time['range'] = pd.cut(data_time.position.to_list(), bins)

        phases = data_time.dropna().groupby(['phase','range']).count().unstack('phase')['position'].fillna(0)
        if 'G2/M' not in phases:
            phases['G2/M'] = 0
        if 'G0' not in phases:
            phases['G0'] = 0            
            
        phases['green'] = round(((phases['G1']+phases['G0']) / (phases['G1']+phases['S']+phases['G2/M']+phases['G0']))*100,2)
        phases['magenta'] = round((phases['S'] / (phases['G1']+phases['S']+phases['G2/M']+phases['G0']))*100,2)
        phases['mitosis'] = round((phases['G2/M'] / (phases['G1']+phases['S']+phases['G2/M']+phases['G0']))*100,2)
        #phases = phases.drop(['G1','S'],axis=1)
        phases['position'] = exp_bins
        phases.index = phases['position']
        phases = phases.drop('position',axis=1)
        
        all_times.append(phases)
    all_times = pd.concat(all_times, keys=[0,1,2,3,4,5], names=['day','id'])        
    all_seeds.append(all_times)
all_seeds = pd.concat(all_seeds, keys=np.arange(0,len(files)), names=['tail','day','position'])       
simulations = all_seeds.drop('G2/M',axis=1)
simulations = simulations.drop('G0',axis=1)
simulations_grouped = simulations.reset_index('tail')
bins = np.arange(0,100,20)
simulations_grouped = simulations_grouped.groupby([pd.cut(simulations_grouped['tail'],bins),'day','position']).mean().dropna().drop('tail',axis=1)
```

```python
exp_plots = {}
for day in range(0,6):
    exp_day = experiments.xs(day,level='day').reset_index()
    pos = exp_day['position'].max()-800
    print(pos)
    exp_day = exp_day[exp_day['position']>pos]
    exp_day = exp_day.drop(['tail','position'],axis=1)
    exp_day = exp_day.melt(var_name='signal_type', value_name='signal_value')    
    exp_plots[day] = exp_day
exp_filtered = pd.concat(exp_plots, names=['day','id']).reset_index().drop('id',axis=1)    
```

```python
sim_plots = {}
simulations = simulations_grouped
for day in range(0,6):
    sim_day = simulations.xs(day,level='day').reset_index()
    pos = sim_day['position'].max()-800
    print(pos)
    sim_day = sim_day[sim_day['position']>pos]
    sim_day = sim_day.drop(['tail','position','G1','S'],axis=1)
    sim_day = sim_day.melt(var_name='signal_type', value_name='signal_value')
    sim_plots[day] = sim_day
sim_filtered = pd.concat(sim_plots, names=['day','id']).reset_index().drop('id',axis=1)    
```

```python
from matplotlib import pyplot
a4_dims = (24, 10)
fig, ax = pyplot.subplots(figsize=a4_dims)
sim_filtered['dataset'] = 'simulations'
exp_filtered['dataset'] = 'experiments'
sim_new = sim_filtered[sim_filtered['signal_type']=='green']
exp_new = exp_filtered[exp_filtered['signal_type']=='green']
cdf = pd.concat([sim_new[['day','signal_type','signal_value','dataset']],exp_new[['day','signal_type','signal_value','dataset']]],ignore_index=False, axis=0)    
cdf_green = cdf
palette = {"simulations": "royalblue", "experiments": "seagreen"}
sns.violinplot(data=cdf,x='day', y='signal_value', hue='dataset',ax=ax,width=0.9,palette=palette)
plt.ylim(-15,115)
plt.rcParams.update({'font.size': 24})
plt.savefig('./exp-sim_comparison/green_all_days', dpi=300, bbox_inches='tight')
plt.show()
```

```python
a4_dims = (24, 10)
fig, ax = pyplot.subplots(figsize=a4_dims)
sim_filtered['dataset'] = 'simulations'
exp_filtered['dataset'] = 'experiments'
sim_new = sim_filtered[sim_filtered['signal_type']=='magenta']
exp_new = exp_filtered[exp_filtered['signal_type']=='magenta']
cdf = pd.concat([sim_new[['day','signal_type','signal_value','dataset']],exp_new[['day','signal_type','signal_value','dataset']]],ignore_index=False, axis=0)    
cdf_magenta = cdf
palette = {"simulations": "royalblue", "experiments": "orchid"}
sns.violinplot(data=cdf,x='day', y='signal_value', hue='dataset',ax=ax,width=0.9,palette=palette)
plt.ylim(-15,115)
plt.rcParams.update({'font.size': 24})
plt.savefig('./exp-sim_comparison/magenta_all_days', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Kolmogorov-Smirnov statistic
exp_green = cdf_green[cdf_green["dataset"]=="experiments"]
exp_magenta = cdf_magenta[cdf_magenta["dataset"]=="experiments"]
sim_green = cdf_green[cdf_green["dataset"]=="simulations"]
sim_magenta = cdf_magenta[cdf_magenta["dataset"]=="simulations"]
significance = 0.05
for day in range(0,6):
    print("Day",day)
    expG = exp_green[exp_green["day"]==day]["signal_value"]
    expM = exp_magenta[exp_magenta["day"]==day]["signal_value"]
    simG = sim_green[sim_green["day"]==day]["signal_value"]
    simM = sim_magenta[sim_magenta["day"]==day]["signal_value"]
    
    green   = scipy.stats.ttest_ind(expG,simG,equal_var=False)
    magenta = scipy.stats.ttest_ind(expM,simM,equal_var=False)
    print("Green:",green)
    print("H0:",green[1]>significance)
    print("Magenta",magenta)
    print("H0:",magenta[1]>significance)
```
