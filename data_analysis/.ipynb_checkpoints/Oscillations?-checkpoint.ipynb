```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import scipy
import os
import seaborn as sns 
#from pyabc import (Distribution, History)
```

```python
# Experiments
outgrowth_df = pd.read_csv('./G1S_switchpoint_fitting/'+'outgrowth.csv')
outgrowth_df.set_index(['day', 'tail'], inplace=True)
outgrowth_mean = outgrowth_df.groupby('day').mean()['outgrowth']

percentage_df = pd.read_csv('./G1S_switchpoint_fitting/'+'percentage_100um.csv')
df = percentage_df
for day in range(0,6):
    df.loc[df['day'] == day, 'position'] = (outgrowth_mean[day] - (df.loc[df['day'] == day, 'position']-100)).astype(int)
percentage_df = df
percentage_df.set_index(['day', 'tail', 'position'], inplace=True)
percentage_df = percentage_df.drop(['unlabelled'], axis=1)
experiments = percentage_df
#experiments.drop(['double','mitosis'], axis=1)
```

```python
for day in range(0,6):
    print("Day",day)
    expDay = experiments.xs(day,level="day").reset_index()
    sns.scatterplot(x='position', y='green', data=expDay,style='tail',color='green')
    sns.scatterplot(x='position', y='double', data=expDay,style='tail',color='gray')
    sns.scatterplot(x='position', y='magenta', data=expDay,style='tail',color='magenta')
    plt.legend([],[], frameon=False)
    plt.show()
```

```python
for given_value in np.arange(-1200,100,100):
    posList = {}
    print("Bin:",given_value)
    for day in range(0,6):
        positions = experiments.xs(day,level="day").reset_index()["position"].unique()
        absolute_difference_function = lambda list_value : abs(list_value - given_value)
        closest_value = min(positions, key=absolute_difference_function)
        diff = abs(closest_value - given_value)
        if diff <= 100:
            posList[day] = closest_value
    expPosList = []        
    for pos in posList.values():    
        expPos = experiments.xs(pos,level="position").reset_index().dropna()
        expPosList.append(expPos)
    result = pd.concat(expPosList).groupby("day").mean().drop("tail",axis=1)
    
    sns.scatterplot(data=result, x="day", y="green",color="green")
    sns.scatterplot(data=result, x="day", y="double",color="gray")
    sns.scatterplot(data=result, x="day", y="magenta",color="magenta")
    sns.scatterplot(data=result, x="day", y="mitosis",color="blue")
    plt.show()
    plt.close()
    
    resultAll = pd.concat(expPosList)
    sns.lineplot(data=resultAll, x="day", y="green",color="green",style="tail")
    sns.lineplot(data=resultAll, x="day", y="double",color="gray",style="tail")
    sns.lineplot(data=resultAll, x="day", y="magenta",color="magenta",style="tail")
    sns.lineplot(data=resultAll, x="day", y="mitosis",color="blue",style="tail")
    plt.legend([],[], frameon=False)
    plt.show()
    plt.close()
    
    resultAll = pd.concat(expPosList)
    sns.pointplot(data=resultAll, x="day", y="green",color="green",style="tail")
    sns.pointplot(data=resultAll, x="day", y="double",color="gray",style="tail")
    sns.pointplot(data=resultAll, x="day", y="magenta",color="magenta",style="tail")
    sns.pointplot(data=resultAll, x="day", y="mitosis",color="blue",style="tail")
    plt.legend([],[], frameon=False)
    plt.show()
    plt.close()    
```
