# Recruitment limit model fitting analysis

```python
# required libraries
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from scipy import stats
from pyabc import (Distribution, RV, ABCSMC, DiscreteRandomWalkTransition, AdaptivePopulationSize, History)
```

```python
# Parameters priors
n0min, n0max = 100, 300
lmin, lmax = 500, 1500
taumin, taumax = 1, 192
```

```python
# Simulations database
db_path = ("sqlite:///" + os.path.join("./", "Recruitment_limit_fitting-Results.db"))
h = History(db_path)
run_id = h.id
print("Run id",str(run_id)+":",str(h.n_populations))
```

```python
# Results vizualization
from pyabc.visualization import plot_kde_matrix
dfw = h.get_distribution(m=0)
grid = plot_kde_matrix(*dfw,
                limits={"n0": (n0min, n0max),
                        "tau": (taumin, taumax),
                        "l": (lmin, lmax)})
plt.gcf().set_size_inches(8, 8)
filename = "./" + "Fig_fitA.png"    
plt.savefig(filename, dpi=300)  
```

```python
# Best parameters values
df = dfw[0]
l_mean, l_dt = round(df["l"].mean(),2), round(df["l"].std(),2)
print("l: ", l_mean, "+-", l_dt)
n0_mean, n0_dt = round(df["n0"].mean(),2), round(df["n0"].std(),2)
print("n0: ", n0_mean, "+-", n0_dt)
tau_mean, tau_dt = round(df["tau"].mean(),2), round(df["tau"].std(),2)
print("tau: ", tau_mean, "+-", tau_dt)
```

```python
# Convergence
from pyabc.visualization import plot_acceptance_rates_trajectory,plot_epsilons,plot_credible_intervals
fig = plt.figure()
filename = "./" + "Fig_fitB.png"  
plot_epsilons(h,["Epsilon evolution"]).get_figure().savefig(filename, dpi=300) 
filename = "./" + "Fig_fitC.png"  
plot_acceptance_rates_trajectory(h,["Acceptance rate evolution"]).get_figure().savefig(filename, dpi=300)
```
