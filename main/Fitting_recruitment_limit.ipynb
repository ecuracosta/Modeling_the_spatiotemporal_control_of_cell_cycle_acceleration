# Fitting recruitment limit to switchpoint

```python
# required libraries
import os
import pandas as pd
import numpy as np
from scipy import stats
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt
```

```python
# experimental data
exp_time= [4, 6, 8]  # measured times
exp_sp = [-794, -641, -261]  # experimental switchpoints
exp_err = [-130, -140, -187]  # experimental switchpoints errors
```

```python
root = './simulations/switchpoint/'
os.chdir(root)

n0_array = np.arange(100,305,5)    # Number of cells after amputation
lambda_array = -np.arange(500,1550,50)    # Recruitment limit
tau_array = np.arange(0,4*24+6,6)
cell_diameter = 13.2

heatmap = np.zeros((len(n0_array),len(lambda_array),len(tau_array)))

n0i = 0
for n0 in n0_array:
    li = 0
    for l in lambda_array:
        taui = 0
        for tau in tau_array:
            print('n0:',n0)
            print('lambda:',l)
            print('tau:',tau)              
            df_all = []
            parameters_set = 'n0='+str(n0)+'_l='+str(l)+'_tau='+str(tau)
            files = os.listdir(parameters_set)
            for seed in files:
                opened_file = open(parameters_set+"/"+seed)
                df0 = pd.read_csv(opened_file, delimiter=',')
                recruitment_limit = df0['outgrowth']*cell_diameter
                df_all.append(recruitment_limit)
            df_all = pd.concat(df_all, ignore_index=True, axis=1)
            df_mean = np.asarray(df_all.mean(axis=1))  # theoretical recruitment limit mean among seeds
            ssqi = []
            for switchpoint,error,time in zip(exp_sp,exp_err,exp_time):
            # calculates the value of the error function at each time
                value = (df_mean[time*24]-switchpoint)**2
                ssqi.append(value)
            print('value:', value)
            print('-------')                   
            ssq = sum(ssqi)
            heatmap[n0i,li,taui] = ssq
            taui += 1
        li += 1
    n0i += 1
```

```python
min_value = np.min(heatmap)
```

```python
n0i,li,taui = np.where(heatmap==min_value)
```

```python
n0 = int(n0_array[n0i])
l = int(lambda_array[li])
tau = int(tau_array[taui])
print("Best fit parameters values")
print('n0:',n0)
print('l:',l)
print('tau:',tau)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax = sns.heatmap(heatmap[n0i[0],:,:], annot=False, fmt="g", cmap=sns.cubehelix_palette(90, start=2, rot=0.05, dark=0.2, light=1, reverse=True),xticklabels=tau_array,yticklabels=lambda_array,cbar=True)
cbar = ax.collections[0].colorbar
cbar.set_label('Residual sum of squares', labelpad=32)
plt.ylabel(r'$\lambda$' + ' (' + r'$\mu$'+'m)')
plt.xlabel(r'$\tau$' + ' (' + r'$hs$'+')')
plt.suptitle('Parameter space fixing '+r'$N_0$'+' = '+str(n0), y=0.98, x=0.42, size=65)
fig.set_size_inches(18.5, 12.5)
sns.set(font_scale=1.5)
filename = "../figures/" + "Fig_2A1.png"
fig.savefig(filename, dpi=300, bbox_inches='tight')
plt.show()
```
